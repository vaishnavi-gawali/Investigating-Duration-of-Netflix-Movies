```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the netflix_data file
netflix_df = pd.read_csv('netflix_data.csv')

# Create a subset by removing TV shows and keeping just movies
netflix_subset = netflix_df[netflix_df['type'] != "TV Show"]

# Create a subset by to investigate the movie duration using the relevant columns
netflix_movies = netflix_subset[['title','country','genre','release_year','duration']]

#Create a subset with movies less than 60 minutes
short_movies = netflix_movies[netflix_movies['duration']<60]
```

```python
#Assign colors to different genres
colors = []
for index,row in netflix_movies.iterrows():
    genre1 = row['genre']
    if genre1 == "Children":
        colors.append("Red")
    elif genre1 == "Documentaries":
        colors.append("Yellow")
    elif genre1 == "Stand-Up":
        colors.append("Blue")
    else:
        colors.append("Green")
```

```python
#use scatterplot to find if the movie duration is getting shorter

fig = plt.figure(figsize=(12,8))

plt.scatter(netflix_movies.release_year,netflix_movies.duration,c=colors)
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.title("Movie Duration by Year of Release")
plt.show()
```
